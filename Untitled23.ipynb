```python
# ✅ Step 1: Import Libraries
import pandas as pd
import numpy as np

# ✅ Step 2: Upload the Dataset
from google.colab import files
uploaded = files.upload()  # Upload the Uber dataset CSV file

# ✅ Step 3: Load the Dataset
df = pd.read_csv(list(uploaded.keys())[0])
print("✅ Dataset Loaded Successfully")
print("Shape of dataset:", df.shape)

# ✅ Step 4: Quick Overview of Data
print("\nFirst 5 rows:")
print(df.head())

print("\nData Info:")
print(df.info())

print("\nSummary Statistics:")
print(df.describe(include='all'))

# ✅ Step 5: Check for Missing Values
print("\nMissing Values in Each Column:")
print(df.isnull().sum())

# Fill missing values (example: fill numeric with mean, categorical with mode)
for col in df.columns:
    if df[col].dtype == 'object':
        df[col].fillna(df[col].mode()[0], inplace=True)
    else:
        df[col].fillna(df[col].mean(), inplace=True)

print("\nMissing Values After Cleaning:")
print(df.isnull().sum())

# ✅ Step 6: Check for Duplicates
duplicates = df.duplicated().sum()
print("\nNumber of Duplicate Rows:", duplicates)

# Remove duplicates
df.drop_duplicates(inplace=True)

print("Shape after removing duplicates:", df.shape)

# ✅ Step 7: Handle Incorrect Data Types (Example: Date columns)
# If 'Date/Time' column exists, convert it to datetime
if 'Date/Time' in df.columns:
    df['Date/Time'] = pd.to_datetime(df['Date/Time'], errors='coerce')

print("\nData Types After Conversion:")
print(df.dtypes)

# ✅ Step 8: Handle Inconsistent Values (Example: lowercase city names)
if 'City' in df.columns:
    df['City'] = df['City'].str.strip().str.lower()

# ✅ Step 9: Detect & Remove Outliers (Example: Trip Duration)
if 'Trip_Duration' in df.columns:
    Q1 = df['Trip_Duration'].quantile(0.25)
    Q3 = df['Trip_Duration'].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    df = df[(df['Trip_Duration'] >= lower_limit) & (df['Trip_Duration'] <= upper_limit)]

print("\nShape after removing outliers:", df.shape)

# ✅ Step 10: Save Cleaned Data
df.to_csv("uber_cleaned.csv", index=False)
print("\n✅ Cleaned dataset saved as uber_cleaned.csv")

# ✅ Display final cleaned data info
print("\nFinal Cleaned Data Info:")
print(df.info())
```
